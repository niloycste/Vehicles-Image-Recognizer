```python
#!export
model = load_learner(f'models/vehicle-recognizer-v2.pkl')
```

```python
#!export
vehicle_labels = (
    "car",
    "motorcycle",
    "bicycle",
    "truck",
    "bus",
    "van",
    "rickshaw",
    "scooter",
    "skateboard",
    "ambulance",
    "fire truck",
    "tractor",
    "segway",
    "unicycle",
    "jet ski",
    "helicopter",
    "airplane",
    "boat",
    "kayak",
    "hovercraft"
)

def recognize_image(image):
  pred, idx, probs = model.predict(image)
  return dict(zip(vehicle_labels, map(float, probs)))
```

```python


img = PILImage.create(f'test_images/bus.jpg')
img.thumbnail((192,192))
img

```

```python
recognize_image(img)
```

```python
#!export
import gradio as gr


image_input = gr.Image()
label_output = gr.Label()

examples = [
    'test_images/bus.jpg',
    'test_images/car.jpg',
    'test_images/helicopter.jpg',
    'test_images/plane.jpg'
]

iface = gr.Interface(fn=recognize_image, inputs=image_input, outputs=label_output, examples=examples)
iface.launch(inline=False)
```
